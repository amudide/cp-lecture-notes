```python
def factorial_r(n):
    
    # base case
    if n == 1:
        return 1
    
    return n * factorial_r(n - 1)
    

    
    
# factorial_r(4) = print(4 * factorial_r(3)) = print(4 * 3 * factorial_r(2)) = print(4 * 3 * 2 * 1) = print(24)


print(factorial_r(5))
```

```python
def factorial_i(n):
    # output the factorial of n.   n!.  n * (n-1) * .. * 3 * 2 * 1
    
    product = 1
    
    for i in range(1, n + 1):     # i = 1, i = 2, i = 3,.... i = n
        product = product * i
        # product = product * n
    
    print(product)
```

```python
factorial_i(4)
```
